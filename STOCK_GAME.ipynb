```python
# Importowanie bibliotek
import plotly.graph_objects as go
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
from datetime import datetime
import datetime as dt
import pandas_ta as ta
import numpy as np
import plotly.express as px
import random
from pprint import pprint 
pd.options.mode.chained_assignment = None # default='warn'

#Pobranie danych

company=["ETH-USD"] # Wybór waluty, dane z https://finance.yahoo.com/  

start= dt.datetime(2000,2,1) #początek danych
end=dt.datetime(2025,9,6) #koniec danych

yf.pdr_override() 
df=pdr.get_data_yahoo(company, start, end) 
df.reset_index(inplace=True)

#Obliczanie średnich kroczących
s
df['10'] = ta.sma(df["Close"], length=10) 
df['20'] = ta.sma(df["Close"], length=20)
df['30'] = ta.sma(df["Close"], length=30) 
df['50'] = ta.sma(df["Close"], length=50) 
df['100'] = ta.sma(df["Close"], length=100) 

#Losowanie przedziału startowego z danych historycznych. 

start = random.randrange(1,len(df)-50)
end = start + 50

buy = 0 #ostatnie kupno
sell = 0 #ostatnia sprzedaż
short = 0 #ostatni short
rebuy = 0 #ostatni koniec shorta
profits = [] #lista wyników tranzakcji
```

```python
# Wyświetla cały wykres.

sd = df[:end] #skraca df do wybranej, losowej daty (od tej daty rozpoczynamy zabawe w inwestora)
fig2 = go.Figure(
    [ go.Scatter(
            x=sd["Date"],
            y=sd["50"],
            yaxis="y2",
            name="SMA_50"
        ),
     go.Scatter(
            x=sd["Date"],
            y=sd["100"],
            yaxis="y2",
            name="SMA_100"
        ),

    go.Candlestick(
        x=sd["Date"],
        open=sd["Open"],
        high=sd["High"],
        low=sd["Low"],
        close=sd["Close"],
        yaxis="y2",
        name="Candlestick"
        ),
    ]
).update_layout(xaxis_rangeslider_visible=False)
fig2.show()
```

```python
#Zapytanie o fundusz początkowy oraz fee (w %)
validation=0 #zmienna walidacyjna pytań do inwestora

while validation==0:
    try:
        capital = float(input("Jakim funduszem chcesz grać: "))
        validation=1
    except:
        print("Nieprawidlowa wartosc")
        continue

validation=0 #zmienna walidacyjna pytań do inwestora

while validation==0:
    try:
        fee = float(input("Jakie jest % fee na twojej giełdzie (taker): "))*0.01
        validation=1
    except:
        print("Nieprawidlowa wartosc")
        continue

cumulative_profit = [capital] #obliczanie posiadanego kapitału
starting_money = capital #Zapisuje fundusz początkowy


#Tworzę data frame słuzący do zapisywania kupn i sprzedaży
save_sd = df[start:]   
save_sd['buys'] = np.nan
save_sd['sells'] = np.nan
save_sd['shorts'] = np.nan
save_sd['rebuys'] = np.nan
save_sd.reset_index(inplace=True, drop=True)
save_sd_index = 0

validation=0 #zmienna walidacji pytania o intencję inwestora


```

```python
game = '1' #Zmienna wyrażająca chęć kontynuowania gry

while game!= 5:  #Jeżeli nie zakończyliśmy gry
    validation=0 #zmienna walidacji pytania o intencję inwestora
    
    
    #Tworzę 50 dniowy wycinek danych
    sd = df[start:end] 
    sd.reset_index(inplace=True, drop=True)



    #Tworzę wykres dla 50 dni
    fig = go.Figure(
        [
            go.Scatter(
                x=sd["Date"],
                y=sd["10"],
                yaxis="y2",
                name="SMA_10",
            ),
            
            go.Scatter(
                x=sd["Date"],
                y=sd["20"],
                yaxis="y2",
                name="SMA_20",

            ),
            
            go.Scatter(
                x=sd["Date"],
                y=sd["30"],
                yaxis="y2",
                name="SMA_30",

            ),
            
            go.Scatter(
                x=sd["Date"],
                y=sd["50"],
                yaxis="y2",
                name="SMA_50",

            ),
             
            go.Scatter(
                x=sd["Date"],
                y=sd["100"],
                yaxis="y2",
                name="SMA_100",

            ),
            
            go.Candlestick(
                x=sd["Date"],
                open=sd["Open"],
                high=sd["High"],
                low=sd["Low"],
                close=sd["Close"],
                yaxis="y2",
                name="Candlestick"
            ),
            
            go.Scatter(
                x = save_sd["Date"][save_sd_index : save_sd_index+50],
                y = save_sd["buys"][save_sd_index : save_sd_index+50],
                mode = 'markers',
                marker=dict(
                    color='darkgreen',
                    size=10),
                yaxis="y2",
                name="buys"
            ),
            
            go.Scatter(
                x = save_sd["Date"][save_sd_index : save_sd_index+50],
                y = save_sd["sells"][save_sd_index : save_sd_index+50],
                mode = 'markers',
                marker=dict(
                    color='darkred',
                    size=10),
                yaxis="y2",
                name="sells"
            ),
                        
            go.Scatter(
                x = save_sd["Date"][save_sd_index : save_sd_index+50],
                y = save_sd["shorts"][save_sd_index : save_sd_index+50],
                mode = 'markers',
                marker=dict(
                    color='crimson',
                    size=10),
                yaxis="y2",
                name="shorts"
            ),

            go.Scatter(
                x = save_sd["Date"][save_sd_index : save_sd_index+50],
                y = save_sd["rebuys"][save_sd_index : save_sd_index+50],
                mode = 'markers',
                marker=dict(
                    color='darkseagreen',
                    size=10),
                yaxis="y2",
                name="rebuys"
            )


        ]
    ).update_layout(xaxis_rangeslider_visible=False)
    fig.show()
    
    if buy!=0 :
        print("Obecny stan tranzakcji long: ", round(((sd['Close'][49]/buy - 1)*100 - (100*fee)) * levar,2), "%")    
    
    if short!=0:
        print("Obecny stan tranzakcji short: ", round(((short/sd['Close'][49] - 1)*100 - (100*fee)) * levar,2), "%")


    
    #Pytanie o intencje inwestora
    while validation==0:
        try:
            answer = int(input("Co robisz?\n Kupujesz(0),shortujesz(3), nic(1), kończysz tranzakcję(2),  koniec gry(5): "))

            if answer==2 and buy==0 and short==0:
                print("Nie możesz zakończyć tranzakcji - jeszcze żadnej nie utworzyłeś")

            elif (answer==0 or answer==3) and (buy!=0 or short!=0):
                print("Nie możesz zlecić kolejnej tranzakcji przed zakończeniem poprzedniej")

            elif answer not in [0, 1, 2, 3, 5]:
                print("Nieodpowiednia wartosc")
                       
           
            
            else:
                validation=1
                
        except: 
            print("Nieodpowiednia wartosc")
    

    
    #Wynik intencji inwestora
    if answer==0 and buy==0 and sell==0: #Zakup 
        validation = 0
        
        while validation == 0 :
            try:
                invest = int(input("Ile chcesz zainwestowac: "))
                levar = int(input("Jaka dzwignia: "))

                print('Kupiles aktywo w cenie: ', sd['Close'][49]) 
                buy = sd['Close'][49] * (1 -fee)
                save_sd['buys'][49 + save_sd_index] = sd['Close'][49] 
                validation =1
            except:
                print("Nieodpowiednia wartosc...")
                continue
        
    elif answer==1:
        
        print("Nic nie robisz...")
        
    elif answer==2 and buy!=0 and sell==0 :
        print("Sprzedales aktywo za: ", sd['Close'][49])
        sell = sd['Close'][49]
        
        profit = ((sell/buy - 1)- fee)*levar * invest
        
        profits.append(profit)
        cumulative_profit.append(cumulative_profit[-1]+profit)
        
        save_sd['sells'][49 + save_sd_index] = sd['Close'][49]
        
        if profit>=0:
            print("Gratuluje, zarobiles: ", profit)

        
        else:
            print("Straciles: ", profit)
     
        buy = 0
        sell = 0
        
    
    elif answer==3 and short==0 and rebuy==0: #Short 
        validation = 0
        
        while validation == 0 :
            try:
                invest = int(input("Ile chcesz zainwestowac: "))
                levar = int(input("Jaka dzwignia: "))

                print('Shortujesz aktywo w cenie: ', sd['Close'][49]) 
                short = sd['Close'][49] * (1 + fee)
                save_sd['shorts'][49 + save_sd_index] = sd['Close'][49] 
                validation =1
            except:
                print("Nieodpowiednia wartosc...")
                continue



    elif answer==2 and short!=0 and rebuy==0 :
        print("Sprzedales aktywo za: ", sd['Close'][49])
        rebuy = sd['Close'][49]
        
        profit = ((short/rebuy - 1)- fee)*levar * invest
        
        profits.append(profit)
        cumulative_profit.append(cumulative_profit[-1]+profit)
        
        save_sd['rebuys'][49 + save_sd_index] = sd['Close'][49]
        
        if profit>=0:
            print("Gratuluje, zarobiles: ", profit)

        
        else:
            print("Straciles: ", profit)
     
        short = 0
        rebuy = 0

    elif answer==5:
        game = 5
        print("Koniec gry!")
        
    #Dodanie 1 do początku i końca przedziału 50 dni
    start = start + 1
    end = end + 1
    

    #Raport 
    print("Posiadasz obecnie: ", cumulative_profit[-1])
    print("Inwestujesz od: ", save_sd_index, "dni.")
    zysk = cumulative_profit[-1] - starting_money
    if zysk>=0:
        print("Zarobiles jak dotad: ", round(zysk,2))

    else:
        print("Straciles jak dotad: ", round(zysk,2))
        
    #Dodanie 1 do indexu zapisującego 
    save_sd_index = save_sd_index + 1 
```
